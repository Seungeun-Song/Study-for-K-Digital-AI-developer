```
import numpy as np

def Machine(x, w, b):
  y_hat = (w*x) + b
  return y_hat
```

```
x = np.array([1,3,5,7,9])
w=2
b=1
```

```
Machine(x,w,b)
```

```
def Gradient(x, y, w, b):
  y_hat = Machine(x, w, b)

  # 기울기 = 편미분
  
  dw = np.mean((y-y_hat)*(-2*x))
  db = np.mean((y-y_hat)*(-2))

  return dw, db
```

```
y= np.array([2,4,6,8,10])

dw, db = Gradient(x,y,w,b)
```

```
print('dw is ', dw)
print('db is ', db)
```

```
def Learning(x,y,w,b,step):
  dw, db = Gradient(x,y,w,b)

  # 편미분 * 학습률
  uw = w - step*dw
  ub = b - step*db

  return uw, ub
```

```
step= 0.05

uw, ub = Learning(x,y,w,b,step)
```

```
print('Updated_w is ', '%.3f' %uw)
print('Updated_b is ', '%.3f' % ub)
```

```
nw, nb = Gradient(x,y,uw,ub)
```

```
print('nw is ', nw)
print('nb is ', nb)
```

```
def Gradient(x, y, w, b):
  y_hat = Machine(x, w, b)

  dw = np.mean((y-y_hat)*(-2*X))
  db = np.mean((y-y_hat)*(-2))
  Loss = np.mean((y - y_hat)**2)

  return dw, db, Loss
```

```
def Learning(x, y, w, b, step):
  dw, db, Loss =Gradient(x, y, w, b)

  uw = w - step * dw
  ub = b - step * db

  Loss = Loss

  return uw, ub, Loss
```

```
w =2
b= 3
step = 0.001
Error=[]
```

```
for i in range(0,1500):
  uw, ub, Loss = Learning(DATA.inputs, DATA.outputs, w, b, step)

  w = uw
  b = ub
  Error.append(Loss)
```
