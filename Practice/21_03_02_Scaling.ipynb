# Scaling

# 1, 실습데이터 'mpg'

```
import seaborn as sns
DF= sns.load_dataset('mpg')
Cars = sns.load_dataset('Cars')
```

```
DF.info()
```

```
DF.head()
```

## Train_data_set으로 split하지 않아!?

```
X = DF[['weight']]
y = DF['mpg']
```

## 1) without Scaling

### x값 확인

```
X[:5]
```

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9,6))
sns.regplot(x=X, y=y)
plt.xlabel('weight_without_Scaling')
plt.show()
```

## 2) with Normalization

```
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

### 정규화된 x값 확인 ( 0 < X < 1)

```
X_Norm[:20]
```

### 정규화된 x값 모델 생성

```
fig = plt.figure(figsize =(9,6))
sns.regplot(x=X_Norm, y=y)
plt.xlabel('weight_With_Normalization')
plt.show()
```

## 3) with Standardization

```
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

### 표준화된 x값 확인 = 평균인 0을 중심으로 분포

```
X_Stan[:20]
```

### 표준화된 x값 모델 생성

```
fig = plt.figure(figsize =(9,6))
sns.regplot(x=X_Stan, y=y)
plt.xlabel('weight_With_Standardization')
plt.show()
```
