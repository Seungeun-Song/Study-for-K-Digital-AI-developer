```
import seaborn as sns
DF = sns.load_dataset('mpg')
DF.info()
```

```
DF.head()
```

```
X = DF[['weight']]
y = DF[['mpg']]
```

```
X[:5]
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
sns.regplot(X,y, data=DF)
plt.show()
```

```
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

```
X_Norm[:20]
```

```
sns.regplot(X_Norm,y)
plt.show()
```

```
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

```
X_Stan[:20]
```

```
sns.regplot(X_Stan, y)
plt.show()
```

```
plt.figure(figsize=(9,6))
sns.regplot(X_Norm, y)
sns.regplot(X_Stan, y)
plt.xlabel('weight_with_stand_&_Norm')
plt.show()
```

```
from sklearn.model_selection import train_test_split
X = DF[['weight']]
y = DF[['mpg']]
X_train, X_test = train_test_split(X, test_size=0.3, random_state=2045)

scaler = MinMaxScaler()
scaler1 = scaler.fit_transform(X_train)
```

```
scaler.fit(X_train)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```
plt.figure(figsize=(9,6))
plt.plot(X_train_scaled, 'o', label='Train Data')
plt.plot(X_test_scaled, 'x', label = 'Test Data')
plt.plot(scaler1,  '~', label = 'all in one')
plt.show()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, = train_test_split(X, random_state=5, test_size=.1)
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

# X_train - mean(X_train) / std(X_train)
X_train_scaled_ST = scaler.transform(X_train) 

# X_test - mean(X_train) / std(X_train)
X_test_scaled_ST = scaler.transform(X_test)
```

```
fig, axes = plt.subplots(1,2, figsize=(8,6))
plt.scatter(X_train_scaled_ST)
axes[0].scatter(X_test_scaled_ST)
plt.show()
```
