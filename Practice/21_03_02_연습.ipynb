```
import seaborn as sns
DF=sns.load_dataset('mpg')
DF.info()
```

```
DF.head()
```

```
x = DF[['weight']]
y = DF[['mpg']]
```

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9,6))
sns.regplot(x =x, y=y)
plt.xlabel('weight_without_scailing')
plt.show()
```

```
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(x)
```

```
X_Norm[:5]
```

```
fit = plt.figure(figsize=(9,6))
sns.regplot(x=X_Norm, y=y)
plt.xlabel('weight_with_normalization')
plt.show()
```

```
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_stan = scaler2.fit_transform(x)
```

```
fig = plt.figure(figsize=(9,6))
sns.regplot(x=X_stan, y=y)
plt.xlabel('weight_with_standardization')
plt.show()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(x, random_state = 5, test_size=.1)

from sklearn.preprocessing import MinMaxScaler

scaler_N = MinMaxScaler()

# Min(X_train), Max(X_train)
scaler_N.fit(X_train)

# train 데이터의 특징을 가지고 사용할 거라 fit, trainsform을 따로


# X_train - Min(X_train) / Max(X_train) - Min(X_train)
X_train_scaled = scaler_N.transform(X_train)

# X_test - Min(X_train) / Max(X_train) - MIN(X_train)
X_test_scaled = scaler_N.transform(X_test)
```

```
plt.figure(figsize =(9,6))
plt.scatter(X_train_scaled[:,0],X_train_scaled[:,1], 'o', label='훈련 세트', s=60)
plt.scatter(X_test_scaled[:,0],X_test_scaled[:,1], 'x',label='테스트 세트', s=60)
plt.show()
```
