#### 일반적인 웹프로그램 방법론

 : MVC(model_데이터 관점에서 DB를 모델링 --|||--  view_(프론트엔드) UI 부분 설계  --|||--  controller_(백엔드) 웹 어플리케이션과 서버 등의 연결)


#### Django의 방법론

 : MVT(model  --|||--  view_기능 --|||--  template_(프론트엔드))

    모델 코딩하기: 테이블 관련 사항을 개발( models.py, admin.py 파일)
    URLconf 코딩하기 : URL 및 뷰 매핑관계를 정의(urls.py 파일)
    템플릿 코딩하기 : 화면 UI 개발(templates/ 디렉토리 하위의 *.html 파일들)
    뷰 코딩하기 : 애플리케이션 로직 개발(views.py 파일)
